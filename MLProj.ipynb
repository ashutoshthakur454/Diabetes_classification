```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
```

```python
X=df.drop('Outcome',axis=1)
Y=df['Outcome']
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,stratify=Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
classifier.fit(X_train,Y_train)
train_data_accuracy=accuracy_score(classifier.predict(X_train),Y_train)
test_data_accuracy=accuracy_score(classifier.predict(X_test),Y_test)
print(train_data_accuracy,test_data_accuracy)
```

```python
scaler=MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
classifier.fit(X_train_scaled,Y_train)
train_data_accuracy=accuracy_score(classifier.predict(X_train_scaled),Y_train)
test_data_accuracy=accuracy_score(classifier.predict(X_test_scaled),Y_test)
print(train_data_accuracy,test_data_accuracy)
```

```python
neigh=[2,5,8,11,13,15,17]
for i in neigh:
    print("For n=",i)
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    train_data_accuracy=accuracy_score(knn.predict(X_train),Y_train)
    test_data_accuracy=accuracy_score(knn.predict(X_test),Y_test)
    print("Train and test data accuracy: ",train_data_accuracy,test_data_accuracy)

    knn.fit(X_train_scaled,Y_train)
    train_data_accuracy=accuracy_score(knn.predict(X_train_scaled),Y_train)
    test_data_accuracy=accuracy_score(knn.predict(X_test_scaled),Y_test)
    print("Train and test data accuracy on scaled dataset: ",train_data_accuracy,test_data_accuracy)
```

```python
mi=[0.001,0.01,0.1,1,10,100,1000,10000,100000]
for i in mi:
    print("For C=",i)
    clf=LogisticRegression(C=i,max_iter=500)
    clf.fit(X_train,Y_train)
    train_data_accuracy=accuracy_score(clf.predict(X_train),Y_train)
    test_data_accuracy=accuracy_score(clf.predict(X_test),Y_test)
    print("Train and test data accuracy: ",train_data_accuracy,test_data_accuracy)

    clf.fit(X_train_scaled,Y_train)
    train_data_accuracy=accuracy_score(clf.predict(X_train_scaled),Y_train)
    test_data_accuracy=accuracy_score(clf.predict(X_test_scaled),Y_test)
    print("Train and test data accuracy on scaled dataset: ",train_data_accuracy,test_data_accuracy)
```

```python
cparam=[0.01,0.1,1,10,100]
for i in cparam:
    print("C=",i)
    clf = svm.SVC(kernel = 'linear', C=i)
    clf.fit(X_train, Y_train)
    train_data_accuracy=accuracy_score(clf.predict(X_train),Y_train)
    test_data_accuracy=accuracy_score(clf.predict(X_test),Y_test)
    print("Train and test data accuracy: ",train_data_accuracy,test_data_accuracy)

    clf.fit(X_train_scaled,Y_train)
    train_data_accuracy=accuracy_score(clf.predict(X_train_scaled),Y_train)
    test_data_accuracy=accuracy_score(clf.predict(X_test_scaled),Y_test)
    print("Train and test data accuracy on scaled dataset: ",train_data_accuracy,test_data_accuracy)
```

```python
clf=svm.SVC(kernel='poly',degree=3)
clf.fit(X_train, Y_train)
train_data_accuracy=accuracy_score(clf.predict(X_train),Y_train)
test_data_accuracy=accuracy_score(clf.predict(X_test),Y_test)
print("Train and test data accuracy: ",train_data_accuracy,test_data_accuracy)

clf.fit(X_train_scaled,Y_train)
train_data_accuracy=accuracy_score(clf.predict(X_train_scaled),Y_train)
test_data_accuracy=accuracy_score(clf.predict(X_test_scaled),Y_test)
print("Train and test data accuracy on scaled dataset: ",train_data_accuracy,test_data_accuracy)
```

```python
gam=[0.001,0.01,0.1,1,10,100,1000]
see=[0.001,0.01,0.1,1,10,100,1000]
for i in gam:
    for j in see:
        print("For gamma=",i," and C=",j)
        clf=svm.SVC(kernel='rbf',gamma=i,C=j)
        clf.fit(X_train, Y_train)
        train_data_accuracy=accuracy_score(clf.predict(X_train),Y_train)
        test_data_accuracy=accuracy_score(clf.predict(X_test),Y_test)
        print("Train and test data accuracy: ",train_data_accuracy,test_data_accuracy)

        clf.fit(X_train_scaled,Y_train)
        train_data_accuracy=accuracy_score(clf.predict(X_train_scaled),Y_train)
        test_data_accuracy=accuracy_score(clf.predict(X_test_scaled),Y_test)
        print("Train and test data accuracy on scaled dataset: ",train_data_accuracy,test_data_accuracy)
        print("")
```

```python
md=[2,3,4,5,6,7,8]
for i in md:
    print("Max depth:",i)
    clf=DecisionTreeClassifier(max_depth=i)
    clf.fit(X_train, Y_train)
    train_data_accuracy=accuracy_score(clf.predict(X_train),Y_train)
    test_data_accuracy=accuracy_score(clf.predict(X_test),Y_test)
    print("Train and test data accuracy: ",train_data_accuracy,test_data_accuracy)

    clf.fit(X_train_scaled,Y_train)
    train_data_accuracy=accuracy_score(clf.predict(X_train_scaled),Y_train)
    test_data_accuracy=accuracy_score(clf.predict(X_test_scaled),Y_test)
    print("Train and test data accuracy on scaled dataset: ",train_data_accuracy,test_data_accuracy)
    print("")
```
